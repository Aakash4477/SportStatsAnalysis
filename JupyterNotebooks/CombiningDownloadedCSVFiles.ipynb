### Import Libraries

```python
import requests
import pandas as pd
from tqdm.notebook import tqdm
import time
import sys
import logging

import numpy as np

pd.set_option('display.max_columns', 16)
pd.set_option('display.max_rows', None)
```

### Import CSV files to combine

```python
shooting = pd.read_csv('ShootingStats07122020.csv', header=[0,1])
passing = pd.read_csv('PassingStats07122020.csv', header=[0,1])
pass_types = pd.read_csv('PassTypes07122020.csv', header=[0,1])
goalshots = pd.read_csv('GoalShotCreation07122020.csv', header=[0,1])
defensive_action = pd.read_csv('DefensiveAction07122020.csv', header=[0,1])
possession = pd.read_csv('Possession07122020.csv', header=[0,1])
playingtimes = pd.read_csv('PlayingTimes07122020.csv', header=[0,1])
misc_stats = pd.read_csv('MiscStats07122020.csv', header=[0,1])
playersummary = pd.read_csv('PlayerSummary07122020.csv', header=[0,1])
```

```python
playersummary = playersummary[:-3]
# These tables contain players in the squad with no game time.
```

```python
playingtimes = playingtimes[:-3]
```

### Drop columns that aren't needed

```python
cols = [1,2,3]
shooting.drop(shooting.columns[cols],axis=1,inplace=True)
playersummary.drop(playersummary.columns[cols],axis=1,inplace=True)
playingtimes.drop(playingtimes.columns[cols],axis=1,inplace=True)
```

```python
cols = [1,2,3,4]
passing.drop(passing.columns[cols],axis=1,inplace=True)
pass_types.drop(pass_types.columns[cols],axis=1,inplace=True)
goalshots.drop(goalshots.columns[cols],axis=1,inplace=True)
defensive_action.drop(defensive_action.columns[cols],axis=1,inplace=True)
possession.drop(possession.columns[cols],axis=1,inplace=True)
misc_stats.drop(misc_stats.columns[cols],axis=1,inplace=True)
```

### Sort by Name

```python
shooting = shooting.sort_values(by=shooting.columns[0])
passing = passing.sort_values(by=passing.columns[0])
pass_types = pass_types.sort_values(by=pass_types.columns[0])
goalshots = goalshots.sort_values(by=goalshots.columns[0])
defensive_action = defensive_action.sort_values(by=defensive_action.columns[0])
possession = possession.sort_values(by=possession.columns[0])
playingtimes = playingtimes.sort_values(by=playingtimes.columns[0])
misc_stats = misc_stats.sort_values(by=misc_stats.columns[0])
playersummary = playersummary.sort_values(by=playersummary.columns[0])
```

### Reset Index

```python
shooting = shooting.reset_index(drop=True)
passing = passing.reset_index(drop=True)
pass_types = pass_types.reset_index(drop=True)
goalshots = goalshots.reset_index(drop=True)
defensive_action = defensive_action.reset_index(drop=True)
possession = possession.reset_index(drop=True)
playingtimes = playingtimes.reset_index(drop=True)
misc_stats = misc_stats.reset_index(drop=True)
playersummary = playersummary.reset_index(drop=True)
```

### Concat all Dataframes

```python
big_df = pd.concat([shooting,passing,pass_types,goalshots,defensive_action,possession,
                    playingtimes,misc_stats,playersummary], axis=1)
```
